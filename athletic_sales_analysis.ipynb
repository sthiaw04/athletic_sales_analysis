```python
# Import Libraries and Dependencies
import pandas as pd
```

### 1. Combine and Clean the Data
#### Import CSVs

```python
# Read the CSV files into DataFrames.
athletic_sales_2020 = pd.read_csv('Resources/athletic_sales_2020.csv')
athletic_sales_2021 = pd.read_csv('Resources/athletic_sales_2021.csv')
```

```python
# Display the 2020 sales DataFrame
athletic_sales_2020.head()
```

```python
# Display the 2021 sales DataFrame
athletic_sales_2021.tail()
```

#### Check the data types of each DataFrame

```python
# Check the 2020 sales data types.
athletic_sales_2020.dtypes
print(athletic_sales_2020.dtypes)
```

```python
# Check the 2021 sales data types.
athletic_sales_2021.dtypes
print(athletic_sales_2021.dtypes)
```

#### Combine the sales data by rows.

```python
# Combine the 2020 and 2021 sales DataFrames on the rows and reset the index.
combined_sales = pd.concat([athletic_sales_2020, athletic_sales_2021], ignore_index=True)   
combined_sales
```

```python
# Check if any values are null.
combined_sales.isnull().sum()
```

```python
# Check the data type of each column
combined_sales.dtypes
print(combined_sales.dtypes)
```

```python
# Convert the "invoice_date" to a datetime datatype
combined_sales['invoice_date'] = pd.to_datetime(combined_sales['invoice_date'])
```

```python
# Confirm that the "invoice_date" data type has been changed.
combined_sales.dtypes
print(combined_sales.dtypes)
```

### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
# Show the number products sold for region, state, and city.
# Rename the sum to "Total_Products_Sold".
products_sold = combined_sales.groupby(['region', 'state','city']).agg({'product': 'count'}).rename(columns={'product': 'Total_Products_Sold'})

# Show the top 5 results.
products_sold.head()
```

#### Using `pivot_table`

```python
# Show the number products sold for region, state, and city.
pivot_products_sold = combined_sales.pivot_table(index=['region', 'state', 'city'], values='product', aggfunc='count').rename(columns={'product': 'Total_Products_Sold'})

# Rename the "units_sold" column to "Total_Products_Sold"
# did this above

# Show the top 5 results.
pivot_products_sold.head()
```

### 3. Determine which Region had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each region, state, and city.
# Rename the "total_sales" column to "Total Sales"
total_sales = combined_sales.groupby(['region', 'state', 'city']).agg({'total_sales': 'sum'}).rename(columns={'total_sales': 'Total_Sales'})

# Show the top 5 results.
total_sales.head()
```

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each region, state, and city.

pivot_tolal_sales = combined_sales.pivot_table(index=['region', 'state', 'city'], values='total_sales', aggfunc='sum')
# Optional: Rename the "total_sales" column to "Total Sales"


# Show the top 5 results.
pivot_tolal_sales.head()
```

### 4. Determine which Retailer had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each retailer, region, state, and city.
# Rename the "total_sales" column to "Total Sales"
total_sales_retailer = combined_sales.groupby(['retailer', 'region', 'state', 'city']).agg({'total_sales': 'sum'}).rename(columns={'total_sales': 'Total_Sales'})

# Show the top 5 results.
total_sales_retailer.head()
```

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each retailer, region, state, and city.
pivot_total_sales_retailer = combined_sales.pivot_table(index=['retailer', 'region', 'state', 'city'], values='total_sales', aggfunc='sum')

# Optional: Rename the "total_sales" column to "Total Sales"


# Show the top 5 results.
pivot_total_sales_retailer.head()
```

### 5. Determine which Retailer Sold the Most Women's Athletic Footwear

```python
# Filter the sales data to get the women's athletic footwear sales data.
womens_athletic_footwear = combined_sales[
    combined_sales["product"] == "Women's Athletic Footwear"
]
print(womens_athletic_footwear)
```

#### Using `groupby`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
total_womens_athletic_footwear = womens_athletic_footwear.groupby(['retailer', 'region', 'state', 'city']).agg({'product': 'count'}).rename(columns={'product': 'Womens_Footwear_Units_Sold'})

# Show the top 5 results.
total_womens_athletic_footwear.head()
```

#### Using `pivot_table`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
pivot_total_womens_athletic_footwear = womens_athletic_footwear.pivot_table(index=['retailer', 'region', 'state', 'city'], values='product', aggfunc='count').rename(columns={'product': 'Womens_Footwear_Units_Sold'})

# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
# done above

# Show the top 5 results.
pivot_total_womens_athletic_footwear.head
```
